```python
import numpy as np
import pandas as pd
import matplotlib
%matplotlib inline
file="~/github/datasets/598354.csv"
file
data=pd.read_csv(file)
```

```python
select=data[(data.STATION_NAME=='VOLCANO 3.2 N CA US') | (data.STATION_NAME=="TRINIDAD 1.0 ENE CA US")| (data.STATION_NAME=="HETCH HETCHY CA US")]
select1=select[(select.TMIN>-9999)]
```

```python
select
```

```python
stations=select1.groupby("STATION")
```

```python
minmax=stations.agg({'TMIN' : (np.min), 'TMAX' : (np.max)})
```

```python
minmax
```

```python
minmax.plot(kind='bar')
```
